```python
import textwrap

def word_wrap(text, width=120):

    return textwrap.fill(text, width)

# Example usage:
long_text = "This is a long piece of text that needs to be wrapped to fit within a specified width.This is a long piece of text that needs to be wrapped to fit within a specified width."

wrapped_text = word_wrap(long_text, width=120)
print(wrapped_text)
```

```python
import os
from pypdf import PdfReader

# Directory containing PDF files
pdf_directory = "/content/drive/MyDrive/file/"

# List all PDF files in the directory
pdf_file_paths = [os.path.join(pdf_directory, file) for file in os.listdir(pdf_directory) if file.lower().endswith('.pdf')]

# Iterate through each PDF file
for pdf_path in pdf_file_paths:
    # Read the PDF file
    reader = PdfReader(pdf_path)

    # Extract text from each page
    pdf_texts = [p.extract_text().strip() for p in reader.pages]

    # Filter out empty texts
    pdf_texts = [text for text in pdf_texts if text]

    # Print or process the text (e.g., word wrap and print the first page)
    if pdf_texts:
        print(word_wrap(pdf_texts[0]))
    else:
        print(f"No text found in {pdf_path}")
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter, SentenceTransformersTokenTextSplitter
```

```python
character_splitter = RecursiveCharacterTextSplitter(
    separators=["\n\n", "\n", ". ", " ", ""],
    chunk_size=1000,
    chunk_overlap=200
)
character_split_texts = character_splitter.split_text('\n\n'.join(pdf_texts))

print(word_wrap(character_split_texts[10]))
print(f"\nTotal chunks: {len(character_split_texts)}")
```

```python
token_splitter = SentenceTransformersTokenTextSplitter(chunk_overlap=0, tokens_per_chunk=256)

token_split_texts = []
for text in character_split_texts:
    token_split_texts += token_splitter.split_text(text)

print(word_wrap(token_split_texts[10]))
print(f"\nTotal chunks: {len(token_split_texts)}")
```

```python
import chromadb
from chromadb.utils.embedding_functions import SentenceTransformerEmbeddingFunction

embedding_function = SentenceTransformerEmbeddingFunction()
print(embedding_function([token_split_texts[10]]))
```

```python
chroma_client = chromadb.Client()
chroma_collection = chroma_client.create_collection("collect121", embedding_function=embedding_function)

ids = [str(i) for i in range(len(token_split_texts))]

chroma_collection.add(ids=ids, documents=token_split_texts)
chroma_collection.count()
```

```python
query = "what is malaria?"

results = chroma_collection.query(query_texts=[query], n_results=5)
retrieved_documents = results['documents'][0]

for document in retrieved_documents:
    print(word_wrap(document))
    print('\n')
```

```python
def rag(query, retrieved_documents, model="gpt-3.5-turbo"):
    information = "\n\n".join(retrieved_documents)

    #random_query = random.choice(query)

    messages = [
        {
            "role": "system",
            "content": "You are a helpful expert medical helpful teacher bot. Your users want to know what other queries can be asked from the given information."
            "You will be shown the user's question, and the relevant information from the medical course. Generate 5 different queries the user's can get information from the given inforamtion and to you using only this information ."

        },
        {"role": "user", "content": f"Question: {query}. \n Information: {information}"}
    ]

    response = openai_client.chat.completions.create(
        model=model,
        messages=messages,
    )
    content = response.choices[0].message.content
    return content
```

```python
output = rag(query=query, retrieved_documents=retrieved_documents)

print(word_wrap(output))
```

```python

# it is for providing answer in bullet point
#provide top 4 point
#provide 5 similar question

def answer(query, retrieved_documents, model="gpt-3.5-turbo"):
    information = "\n\n".join(retrieved_documents)

    messages = [
        {
            "role": "system",
            "content": "You are a helpful expert medical bot. Your users are asking questions about information contained in a medical workers course. Provide answer in bullet points."
            "You will be shown the user's question, and the relevant information from the medical course. Answer the user's question using only this information."
        },
        {"role": "user", "content": f"Question: {query}. \n Information: {information}"}
    ]

    response = openai_client.chat.completions.create(
        model=model,
        messages=messages,
    )
    content = response.choices[0].message.content

    # Process the content and format it into point-wise with each point on a new line
    content_points = [f"{i+1}. {line}" for i, line in enumerate(content.split('\n')) if line.strip()]
    content_points = content_points[:5]

    return "\n".join(content_points)
```

```python
output = rag(query=query, retrieved_documents=retrieved_documents)

print(word_wrap(output))
```
